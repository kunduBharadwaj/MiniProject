```python
chroma_client = chromadb.Client()
```

```python
client = Groq(api_key=userdata.get('groq_api'))
```

```python
emb_fn = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
```

```python
try:
  chroma_client.delete_collection("Hr_hand_book")
except:
  pass
```

```python
collection  = chroma_client.create_collection(name="Hr_hand_book", embedding_function=emb_fn)
```

```python
raw_chunks = raw_handbook_text.split("##")
```

```python
clean_chunks = []

for chunk in raw_chunks:
    chunk = chunk.strip()
    if len(chunk) > 0:
      chunk = chunk.replace("\n", " ")
      clean_chunks.append(chunk)
```

```python
clean_chunks
```

```python
chunk_id = [f'chunk_{i}' for i in range(len(clean_chunks))]
```

```python
collection.add(
    ids=chunk_id,
    documents=clean_chunks
)
```

```python
def hr_bot(user_question):
    print(f'Let me explain about {user_question}')
    result =  collection.query(query_texts=[user_question],n_results=1)
    retrieved_content = result['documents'][0][0]
    print(f'ğŸ” System Retrieved: ...{retrieved_content}')
    system_prompt = f"""
        You are an HR Assistant.
        You must answer the user's question using ONLY the provided Context below.
        If the answer is not in the context, say "I don't know based on the handbook."
        Do NOT invent information.
    Context:
    {retrieved_content}

        """
    completion = client.chat.completions.create(
        model = 'llama-3.1-8b-instant',
        messages = [
            {"role": "system", "content": system_prompt},
            {"role":"user", "content": user_question}
        ],
        temperature = 0 # Good choice of temperature
    )

    #print(completion)
    answer = completion.choices[0].message.content # Try removing and printing you will understand why we need this!
    print(f"ğŸ¤– HR Bot: {answer}\n" + "-"*40) # Cosmetic Print statement
```

```python
hr_bot("I want to resign?")
```
